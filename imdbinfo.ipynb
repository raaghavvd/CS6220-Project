```python
import csv
import json
import tqdm
movie_dict=dict()
unsuable=0
# with open('/Users/raaghav/spark-stuff/processedMovids.csv', encoding="ISO-8859-1") as csv_file:
for i,y in tqdm.tqdm(zip(movie_id,zipped_list)):
    # if line_count == 0:
        # print(row)r
        
    
        title_year=y[1]
        movie_tile=y[0]
        # if int(row[0])>6391:
        
            

        # print(title_year)
        # print(movie_tile)
        url='http://www.omdbapi.com/?apikey=e15fb59d'

        PARAM={'t':movie_tile,'y':title_year}
        try:
            r = requests.get(url = url, params = PARAM)
            data=r.json()
        except:
            print(movie_tile)
            continue

        # print(data)
        # print(data.keys())
        if 'Error' in data.keys():
            unsuable+=1
            continue

        #Extract RunTime
        runtime=data['Runtime']
        runtime=re.findall(r'\b\d+\b', runtime)
        # print(runtime)
        try:
            runtime=runtime[0]
        except:
            continue
        # print
    
        #Wins
        if data['Awards']=='N/A':
            awards=0

        else:
            awards=1
        # print (awards)

        #Metascore
        metaScore=data['Metascore']
        # print(metaScore)


        #IMDBRatings
        imdbScore=data['imdbRating']
        # print(imdbScore)

        #Language
        if 'English' in  data['Language']:
            lang=1
        else:
            lang=0
        


        #Votes
        votes=data['imdbVotes']
        votes = votes.replace(',', '')

    
        

        #Plot
        plot=data['Plot']

        movie_dict[str(i)]={'Movie': movie_tile, 'Runtime': runtime,'IMDB Ratings': imdbScore,'IMDB Votes':votes,'Plot':plot,'Awards':awards,'Metascore':metaScore,'Language':lang}
        
        # if i==2:
        #     break
        # i+=1
        # break
    #     print(f'Column names are {", ".join(row)}')
    #     line_count += 1
    # else:
    #     print(f'\t{row[0]} works in the {row[1]} department, and was born in {row[2]}.')
    #     line_count += 1
    # print(f'Processed {line_count} lines.')
# json_object = json.dumps(movie_dict, indent = 4) 
with open('movieLens_1M_dump.txt', 'w') as outfile:
    json.dump(movie_dict, outfile,indent=4)
```
